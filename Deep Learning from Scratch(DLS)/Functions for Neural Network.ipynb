```python
# 계단함수의 그래프 그리기
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype = np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 시그모이드 함수 구현하기
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)  # 계단함수와 다른 점은 출력값 y 가 sigmoid 함수에 의한 것.
plt.plot(x, y)
plt.show()
```

```python
# ReLU(Rectified Linear Unit) 함수 (활성화 함수로 최근 자주 사용)
def relu(x):
    return np.maximum(0, x)

y = relu(x)
plt.plot(x,y)
plt.show()
```
